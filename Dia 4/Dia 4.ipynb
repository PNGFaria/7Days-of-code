```python
CDU = list()
for c in dfcompleto['localizacao']:
    if 000 <= c <= 99: 
        CDU.append('Generalidades. Ciência e conhecimento.')
    elif 100 <= c <= 199: 
        CDU.append('Filosofia e psicologia.')
    elif 200 <= c <= 299: 
        CDU.append('Religião.')
    elif 300 <= c <= 399: 
        CDU.append('Ciências sociais.')
    elif 400 <= c <= 499: 
        CDU.append('Classe vaga. Provisoriamente não ocupada.')
    elif 500 <= c <= 599: 
        CDU.append('Matemática e ciências naturais.')
    elif 600 <= c <= 699: 
        CDU.append('Ciências aplicadas.')
    elif 700 <= c <= 799: 
        CDU.append( 'Belas artes.')
    elif 800 <= c <= 899: 
        CDU.append('Linguagem. Língua. Linguística.')
    elif 900 <= c <= 999: 
        CDU.append('Geografia. Biografia. História.')
```

```python
def frequencias(vari):
    d = pd.DataFrame(dfcompleto.groupby(vari).size().sort_values(ascending=False))
    resu = ((d[0].values/d[0].sum())*100).round(1)
    d.rename(columns={0: 'Quantidade'}, inplace= True)
    d['Frequencia'] = resu
    return d
```
