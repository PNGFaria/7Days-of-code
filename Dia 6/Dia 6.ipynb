```python
def frequencias(vari):
    d = pd.DataFrame(dfcompleto.groupby(vari).size().sort_values(ascending=False))
    resu = ((d[0].values/d[0].sum())*100).round(1)
    d.rename(columns={0: 'Quantidade'}, inplace= True)
    d['Frequencia'] = resu
    return d
```
